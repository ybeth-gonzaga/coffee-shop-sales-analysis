### Fetch the datasets

```python
import pandas as pd

df = pd.read_csv("clean_coffee_shop_sales.csv")
df_monthly = pd.read_csv("monthly_sales_summary.csv")
```

```python
df_monthly.head()
```

```python
df.info()
```

### Calculate the total number of orders for each respective month.

```python
# Initialize an empty list to store order counts
qty_sold = []

# Iterate over each row in df_monthly
for index, row in df_monthly.iterrows():
    # Get the sum of quantity sold for the current month
    sum = df[df['transaction_month'] == row['transaction_month']]['transaction_qty'].sum()
    
    # Append the count to the orders_count list
    qty_sold.append(sum)

# Assign the orders_count list as a new column in df_monthly
df_monthly['qty_sold'] = qty_sold

# Display the updated DataFrame
print(df_monthly)
```

### Calculate the difference in the number of orders between the selected month and the previous month.

```python
df_monthly['qty_sold_difference'] = df_monthly['qty_sold'].diff()
df_monthly
```

### Determine the month-on-month increase or decrease in the number of orders.

```python
df_monthly['qty_sold_percent_difference'] = df_monthly['qty_sold'].pct_change()*100
df_monthly
```
